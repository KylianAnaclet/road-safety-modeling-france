```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
# 1. Chargement des données
# Remplacez 'votre_fichier.csv' par le nom réel de votre fichier
df = pd.read_csv('accidents_clean.csv', sep=';', decimal=',')
```

```python
# 2. Vérification et nettoyage rapide
# On s'assure d'avoir les colonnes nécessaires (ex: 'lat', 'long', 'score_severite')
df = df.dropna(subset=['lat', 'long', 'Score_Severite'])
df = df[df['mois'].isin([6, 7, 8])]
```

```python
# 3. Création de la carte centrée sur la France
carte_france = folium.Map(location=[46.603354, 1.888334], zoom_start=6)

# 4. Préparation des données pour la HeatMap
# Format attendu : liste de [lat, long, poids(score)]
heat_data = df[['lat', 'long', 'Score_Severite']].values.tolist()

# 5. Ajout de la couche HeatMap
HeatMap(heat_data, radius=10, blur=15, max_zoom=1).add_to(carte_france)

# 6. Sauvegarde
carte_france.save("heatmap_accidents_severite.html")
```
